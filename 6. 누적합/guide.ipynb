# 누적합

누적합 문제는 사실상 DP로 분류된다.

dp[i] : i 이전까지 모든 수의 합

을 저장하는 dp를 만든다.

이후 i~j까지 합을 원하면 dp[j] - dp[i-1]을 하면 원하는 값을 얻을 수 있다.

(dp[j] = 0~j까지 모든 수의 합, dp[i-1] = 0~i-1까지 모든 수의 합)

# Template - 누적합

```python
num_list = [1, 2, 3, 4, 5, 6]
dp = [0]

for num in num_list:
    dp.append(dp[-1] + num)

i = 2
j = 4
print(dp[j] - dp[i-1]) # 2 + 3 + 4 = 9
```

# Template - 2차원 누적합

"구간합 구하기 5" 문제가 해당 template를 적용하는 문제.

2차원 구간 합도 간간히 출제되는 dp 문제이다.



D = B + C - A



(1, 1) ~ (2, 2) 까지 합 = 19 - 5 - 6 + 1

```python
N = 4
board = [
    [1, 2, 3, 4],
    [2, 3, 4, 5],
    [3, 4, 5, 6],
    [4, 5, 6, 7]
]

# dp[i][j] : (0, 0) 부터 (i, j)까지 합
dp = [[0] * (N+1)]

# dp 만들기
for row in range(N):
    temp_row = [0]
    for col in range(N):
        cur_num = board[row][col]
        cur_sum_num = temp_row[-1] + dp[-1][col+1] + cur_num - dp[-1][col]
        temp_row.append(cur_sum_num)
    dp.append(temp_row)

# (x1, y1) ~ (x2, y2) 구간합
x1, y1, x2, y2 = 2, 2, 3, 4
answer = dp[x2][y2] + dp[x1-1][y1-1] - dp[x1-1][y2] - dp[x2][y1-1]
print(answer)
```
