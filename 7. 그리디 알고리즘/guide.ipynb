# 그리디 알고리즘

그리디(greedy)는 '욕심 많은' 이란 뜻이다.

그리디 알고리즘은 미래의 선택은 알빠아니고 현재의 선택에서 가장 큰 이득을 취하는 선택을 하는 문제들이다.

즉, 무식하게 가장 좋은 선택들을 하는 알고리즘을 만들면 된다.
